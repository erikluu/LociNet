```python
import os
import pandas as pd
```

```python
def directory_to_dataframe(root_dir):
    file_data = []

    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                relative_path = os.path.relpath(file_path, root_dir)

                dir_path = os.path.dirname(relative_path)

                with open(file_path, 'r') as f:
                    file_text = f.read()

                # Append the file data to the list
                file_data.append({
                    'filename': file,
                    'path': dir_path,
                    'file_text': file_text,
                    'combined': f'Title: {file} Content: {file_text}'
                })

    return pd.DataFrame(file_data)
```

```python
files_df = directory_to_dataframe('anthonys-second-brain')
files_df
```

```python
with open('output.txt', 'w') as f:
    f.write(files_df.iloc[0].combined.replace('\n', '\\n'))
```

```python
import nltk
from nltk import ngrams
from nltk.tokenize import sent_tokenize

import re
```

```python
# Paragraph Trigrams
def paragraph_tokenizer(text):
    paragraphs = re.split(r'\n+', text)
    res = [p for p in paragraphs if len(sent_tokenize(p)) > 3]

    return res


with open('../data/XAI.txt', 'r') as f:
    paragraphs = paragraph_tokenizer(f.read())


paragraphs
```
