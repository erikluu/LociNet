```python
import os
import openai
import time
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def generate_summary_with_sliding_window(prompt, summary_length=0):
    time.sleep(1)  # Rate limit to 1 request per second
    
    response = openai.Completion.create(
        engine='gpt-3.5-turbo-instruct',
        prompt=prompt,
        max_tokens=200,
        temperature=0.5,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )
    
    summary = response.choices[0].text.strip()
    summary_length += len(summary.split('. '))
    
    if summary_length < 3 or summary_length > 5:
        # Sliding window logic
        prompt = prompt[len(prompt) - 2000:]
        return generate_summary_with_sliding_window(prompt, summary_length)
    
    return summary

summary_prompt = "Generate a concise summary in of the following text entry, focusing on the body and write as if written at a similar readability level. The summary should be 2-3 sentences and highlight the text's main ideas and actions, considering the paper's purpose, markdown elements, and structure employed by the author. Disregard irrelevant links and organize the summary based on the identified structures in the paper. Keep the summary within the token limit."
with open('bee_movie.txt', 'r') as file:
    document = file.read()

summary = generate_summary_with_sliding_window(f'{summary_prompt}: {document}')
summary
```

```python
def generate_summary(prompt):
    response = openai.Completion.create(
        engine='gpt-3.5-turbo-instruct',
        prompt=prompt,
        max_tokens=200,
        temperature=0.5,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )
    return response.choices[0].text.strip()

summary_prompt = "Generate a concise summary in of the following Markdown file, focusing on the title and body as if written at a similar readability level and keeping the same perspective. The summary should be 2-3 sentences and highlight the paper's main ideas and actions, considering the paper's purpose, markdown elements, and structure employed by the author. Disregard irrelevant links and organize the summary based on the identified structures in the paper. Keep the summary within the token limit."
with open('bee_movie.txt', 'r') as file:
    document = file.read()

summary = generate_summary(f'{summary_prompt}: {document}')
summary
```

```python
def generate_label(prompt):
    response = openai.Completion.create(
        engine='gpt-3.5-turbo-instruct',
        prompt=prompt,
        max_tokens=10,
        temperature=0.0, # lower is more deterministic
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )
    return response.choices[0].text.strip()

label_prompt = "Please generate a concise 1-word label that summarizes the content of the markdown file, emphasizing its main focus. The label should be specific and informative without being overly broad. The purpose of the label is to assist in exploration. Imagine a map of a city where more general names are near the top and specific names are near the bottom. FIn"
with open('bee_movie.txt', 'r') as file:
    document = file.read()

label = generate_label(f'{label_prompt}: {document}')
label
```
