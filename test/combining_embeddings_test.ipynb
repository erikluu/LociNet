```python
import sys
sys.path.append('../src')

import embeddings as em
import tqdm
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import torch
```

```python
x = "I am a queen."
y = "I am a princess."
z = "I am a King."
a = "I am Royalty."
b = "Man."
c = "Woman."
d = "Girl."

pca = PCA(n_components=2)

embeddings = em.get_embeddings([x, y, z, a, b, c, d])
em_transformed = pca.fit_transform(embeddings + torch.Tensor([a + b]))

plt.scatter(em_transformed[:, 0], em_transformed[:, 1])
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('PCA Transformed Embeddings')

# Label each point
for i, txt in enumerate([x, y, z, a, b, c, d]):
    plt.annotate(txt, (em_transformed[i, 0], em_transformed[i, 1]))

plt.show()

```
